```python
def rando(x):
    result = ''
    num = random.randint(1, 99)
    if num <= 9:
        result = x + '0'+ '0' + str(num)
    else:
        result = x + '0' + str(num)
        
    return result

df_combine['ID'] = df_combine['ln'].str.upper().str[:3] + '' + df_combine['fn'].str.upper().str[:3] 

ids = df_combine['ID'].values

df_combine['ID'] = df_combine['ID'].apply(lambda x : rando(x))

df_combine['ID'] = df_combine['ID'].apply(lambda x : str.x.replace(x, rando(x)) if x in ids else x)
df_combine.head(50)
```

```python
df_raises = pd.read_csv('/Users/leahsanchez/Workspace/intro_to_python/raises_CAPSTONE.txt', delimiter=',')
df_raises = pd.DataFrame(df_raises)

df_raises = df_raises.applymap(lambda x: str(x).replace("'", ""))
df_raises['raise_amount'] = df_raises['raise_amount'].apply(lambda x: float(x[:-1]))
df_raises['years'] = df_raises['years'].apply(lambda x : float(x))

df_raises.loc[6] = [4.0, 4.0]
df_raises.loc[7] = [6.0, 6.0]
df_raises.loc[8] = [8.0, 8.0]
df_raises.loc[9] = [9.0, 8.0]
df_raises = df_raises.sort_values('years')
df_raises = df_raises.reset_index(drop=True)

for i in range(10, 20):
    df_raises.loc[i] = pd.Series({'years': i + 1, 'raise_amount' : 10.0})
    
df_raises = df_raises.rename(columns={'years': 'Years Employed', 'raise_amount' : 'Raise_Amount'})
df_raises

```

```python
# Create new df with active employees including all columns 

df_active_raises = pd.DataFrame().assign(ID = df_combine['ID'], Name = df_combine['Name'], Gender = df_combine['gender'], DeptCode = df_combine['deptCode'], DeptName = df_combine['dept name'], Hiredate = df_combine['hiredate'], termdate = df_combine['termdate'], SG = df_combine['sg'], Status = df_combine['Status'], Salary = df_combine['salary'] )
df_active_raises['termdate'] = df_active_raises['termdate'].str.strip()
df_active_raises.query("termdate == ''", inplace=True)
df_active_raises.pop('termdate')
df_active_raises
```

```python
# Calc years employed
timestamp = pd.Timestamp(datetime.datetime(2023, 4, 6))
today = timestamp.today()

df_active_raises['Years Employed'] = today - df_active_raises['Hiredate']
df_active_raises['Years Employed'] = (df_active_raises['Years Employed'] / np.timedelta64(1, 'Y')).apply(np.floor)
# Round down to correctly apply the raises
df_active_raises
```

```python
# 2 - The rules for giving a raise are in the file but be careful of the order that your run the raise
# 3 - Based on each employees salary calculate 2 fields 'Raise_Amount' and 'New_Salary'
# check years employed against years column 

df_active_raises['Years Employed'].sort_values()
# 4 years is lowest, 20 years is highest

df_active_raises = pd.merge(df_active_raises, df_raises)
df_active_raises['Raise_Values'] = df_active_raises['Salary'] * df_active_raises['Raise_Amount'] / 100
df_active_raises['New_Salary'] = df_active_raises['Salary'] + (df_active_raises['Salary'] * df_active_raises['Raise_Amount'] / 100)

df_active_raises.sort_values(by='Name')

```
